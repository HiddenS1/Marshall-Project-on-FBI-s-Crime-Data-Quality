```python
import pandas as pd
import numpy as np
```

```python
update = pd.read_csv('https://s3.amazonaws.com/tmp-gfx-public-data/goodstats20230210/ny_updated_participation.csv')
```

```python
update.sample(10)
```

```python
m1 = pd.DataFrame(update['months_reported_in_21'].value_counts()).reset_index()
m2 = pd.DataFrame(update['months_reported_in_22'].value_counts()).reset_index()
compare = pd.merge(m1,m2,on='index',how='outer').rename(columns={'index': 'month'})
```

```python
compare.fillna(0, inplace=True)
```

```python
compare['21_pct'] = compare['months_reported_in_21']/update.shape[0]
compare['22_pct'] = compare['months_reported_in_22']/update.shape[0]
```

```python
compare = compare.sort_values('month').set_index('month')
```

```python
compare = compare.append(compare.iloc[1:11].sum(axis=0), ignore_index=True)
```

```python
compare = compare.drop(compare.index[1:11])
```

```python
compare.reset_index(inplace=True)
compare['index'] = compare['index'].replace({
    0: 'No reporting',
    11: 'Full reporting',
    12: 'Partial reporting'
})
```

```python
compare.rename(columns = {'index': 'Participation'})
```
